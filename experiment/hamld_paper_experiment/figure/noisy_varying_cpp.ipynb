```python
import pandas as pd

# 假设数据存储在名为data.csv的文件中，读取数据
# qldpc_df_acc = pd.read_csv('/home/normaluser/ck/epmld/experiment/epmld_paper_experiment/result/noisy_varying_qldpc_code_acc_results_n_72_r_d.csv')
# qldpc_df_acc = pd.read_csv("/home/normaluser/ck/epmld/experiment/epmld_paper_experiment/result/noisy_varying_qldpc_code_acc_results_20250514_100233.csv")
qldpc_df = pd.read_csv("/home/normaluser/ck/epmld/experiment/epmld_paper_experiment/result/noisy_varying_qldpc_code_acc_cpp_results_20250730_182833.csv")

# print(qldpc_df_acc['nkd'])
# 筛选条件
conditions = (
    (qldpc_df['nkd'] == "[72, 12, 6]") &
    qldpc_df['probability'].isin([0.1, 1, 10,100])
)

# 根据条件筛选数据
qldpc_df = qldpc_df[conditions]


qldpc_df['decoder_method'] = qldpc_df['decoder_method'].replace('EAMLD_cpp', 'EAMLD')

qldpc_df['nkd'] = qldpc_df['nkd'].astype(str)  # 转换为字符串类型
# # 打印结果
# print(result)
```

```python
qldpc_df
```

```python
# 1. 合并X和Z，logical_error_rate取平均值
surface_df_acc = surface_code_df.groupby(['d', 'round', 'probability', 'noise_model','decoder_method', 'have_stabilizer'],
                               as_index=False)['logical_error_rate'].mean()


qldpc_df_acc = qldpc_df.groupby(['nkd', 'round', 'probability', 'noise_model','decoder_method', 'have_stabilizer'],
                             as_index=False)['logical_error_rate'].mean()

surface_code_df['average_per_round_time'] = surface_code_df['total_running_time'] / 1e5
qldpc_df['average_per_round_time'] = qldpc_df['total_running_time'] / 1e5

surface_df_speed = surface_code_df.groupby(['d', 'round', 'probability', 'noise_model','decoder_method', 'have_stabilizer'],
                               as_index=False)['average_per_round_time'].mean()
qldpc_df_speed = qldpc_df.groupby(['nkd', 'round', 'probability', 'noise_model','decoder_method', 'have_stabilizer'],
                             as_index=False)['average_per_round_time'].mean()
```

```python
surface_df_acc_EAMLD = surface_df_acc[surface_df_acc['decoder_method'] == 'EAMLD'].reset_index()
qldpc_df_acc_EAMLD = qldpc_df_acc[qldpc_df_acc['decoder_method'] == 'EAMLD'].reset_index()
surface_df_speed_EAMLD = surface_df_speed[surface_df_speed['decoder_method'] == 'EAMLD'].reset_index()
qldpc_df_speed_EAMLD = qldpc_df_speed[qldpc_df_speed['decoder_method'] == 'EAMLD'].reset_index()
```

```python
surface_df_acc_EAMLD
```

```python
surface_ler_reduction = surface_df_acc.pivot(index='probability', columns='decoder_method', values='logical_error_rate') \
                              .apply(lambda row: 100 * (row['MWPM'] - row['EAMLD']) / row['MWPM'], axis=1)

qldpc_ler_reduction = qldpc_df_acc.pivot(index='probability', columns='decoder_method', values='logical_error_rate') \
                              .apply(lambda row: 100 * (row['BP+OSD'] - row['EAMLD']) / row['BP+OSD'], axis=1)

surface_time_reduction  = surface_df_speed.pivot(index='probability', columns='decoder_method', values='average_per_round_time') \
                              .apply(lambda row: 100 * (row['MWPM'] - row['EAMLD']) / row['MWPM'], axis=1)
                              
qldpc_time_reduction = qldpc_df_speed.pivot(index='probability', columns='decoder_method', values='average_per_round_time') \
                              .apply(lambda row: 100 * (row['BP+OSD'] - row['EAMLD']) / row['BP+OSD'], axis=1)
```

```python
surface_ler_ratio = surface_df_acc.pivot(index='probability', columns='decoder_method', values='logical_error_rate') \
                        .apply(lambda row: row['MWPM'] / row['EAMLD'], axis=1)

qldpc_ler_ratio = qldpc_df_acc.pivot(index='probability', columns='decoder_method', values='logical_error_rate') \
                        .apply(lambda row: row['BP+OSD'] / row['EAMLD'], axis=1)

surface_time_ratio = surface_df_speed.pivot(index='probability', columns='decoder_method', values='average_per_round_time') \
                        .apply(lambda row: row['MWPM'] / row['EAMLD'], axis=1)

qldpc_time_ratio = qldpc_df_speed.pivot(index='probability', columns='decoder_method', values='average_per_round_time') \
                        .apply(lambda row: row['BP+OSD'] / row['EAMLD'], axis=1)
```

```python
# # 转换为LaTeX格式

for i in range(len(surface_df_acc_EAMLD)):
    prob_i = surface_df_acc_EAMLD['probability'][i]
    prob = surface_df_acc_EAMLD['probability'][i] / 10000
    error_rate = surface_df_acc_EAMLD['logical_error_rate'][i]
    avg_time = surface_df_speed_EAMLD['average_per_round_time'][i]
    # 将科学计数法转换为 LaTeX 格式
    prob_str = "{:.1e}".format(prob).replace("e", r"\times10^{") + "}"
    error_rate_str = "{:.3e}".format(error_rate).replace("e", r"\times10^{") + "}"
    ler_reduction = surface_ler_reduction[prob_i]
    time_reduction = surface_time_reduction[prob_i]
    print(f"& ${prob_str}$ & ${error_rate_str}$ & {avg_time:.2f} & {ler_reduction:.2f} & {time_reduction:.2f} \\\\")
```
