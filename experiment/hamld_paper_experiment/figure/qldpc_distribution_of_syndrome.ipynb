### 分布的分析

```python
sample = np.hstack((syndromes, actual_observables))
len_syndrome = syndromes.shape[1]
# 尝试使用 cudf 进行处理
df = cudf.DataFrame(sample)
unique_counts = df.value_counts().reset_index()

# 计算每个syndrome中True的数量
true_counts = unique_counts.iloc[:, :len_syndrome].sum(axis=1)

# 将结果添加到DataFrame中
unique_counts['true_count'] = true_counts

# 按true_count分组统计
grouped_counts = unique_counts.groupby('true_count').sum()

true_counts = grouped_counts.index.values.get()
count_values = grouped_counts.iloc[:, -1].values.get()  # 假设counts在第一列

total = count_values.sum()
frequency_values = count_values / total

# 加权平均汉明权重
true_counts_np = np.array(true_counts)
frequency_values_np = np.array(frequency_values)
average_hw = np.sum(true_counts_np * frequency_values_np) / np.sum(frequency_values_np)
print(f"Average Hamming Weight: {average_hw:.2f}")
```

## Latex 表格

```python
import numpy as np
import cudf
from tabulate import tabulate

nkds = [[72, 12, 6],[144, 12, 12]]
p = 10
noise_model = "si1000"
have_stabilizer = False
related_path = "/home/normaluser/ck/epmld/data/external/epmld_experiment_data/epmld_paper_experiment/overall_performance/qldpc_code"

error_types = ["X", "Z"]

results = []

for nkd in nkds:
    d = nkd[2]
    r_list = [1, d]  # 遍历 r=1 和 r=d
    for r in r_list:
        avg_hw_list = []
        max_hw_list = []

        detector_number = int(nkd[0]/2 + nkd[0] * (r-1))  # syndrome总长度
        
        for error_type in error_types:
            print(f"nkd = {nkd}, r = {r}, error_type = {error_type}")

            dem = generate_qldpc_detector_error_model(
                nkd=nkd, r=r, p=p,
                noise_model=noise_model, error_type=error_type,
                related_path=related_path, have_stabilizer=have_stabilizer
            )
            syndromes, actual_observables = generate_qldpc_syndrome_and_observables(
                nkd=nkd, r=r, p=p,
                noise_model=noise_model, error_type=error_type,
                related_path=related_path, have_stabilizer=have_stabilizer
            )

            sample = np.hstack((syndromes, actual_observables))
            len_syndrome = syndromes.shape[1]

            df = cudf.DataFrame(sample)
            unique_counts = df.value_counts().reset_index()

            true_counts = unique_counts.iloc[:, :len_syndrome].sum(axis=1)
            unique_counts['true_count'] = true_counts

            grouped_counts = unique_counts.groupby('true_count').sum()

            true_counts = grouped_counts.index.values.get()
            count_values = grouped_counts.iloc[:, -1].values.get()

            total = count_values.sum()
            frequency_values = count_values / total

            true_counts_np = np.array(true_counts)
            frequency_values_np = np.array(frequency_values)
            average_hw = np.sum(true_counts_np * frequency_values_np)
            max_hw = true_counts_np.max()

            print(f"  → Average Hamming Weight: {average_hw:.2f}, Max Hamming Weight: {max_hw}\n")

            avg_hw_list.append(average_hw)
            max_hw_list.append(max_hw)

        avg_hw_mean = np.mean(avg_hw_list)
        max_hw_mean = np.mean(max_hw_list)
        ratio = avg_hw_mean / detector_number  # 比值

        results.append({
            "nkd": nkd,
            "r": r,
            "detector_number": detector_number,
            "avg_hw_mean": avg_hw_mean,
            "max_hw_mean": max_hw_mean,
            "ratio": ratio
        })

# 生成 LaTeX 表格
table_data = []
for res in results:
    n, k, d = res["nkd"]
    r = res["r"]
    detector_number = res["detector_number"]
    table_data.append([
        f"$[{n}, {k}, {d}]$",
        f"$r={r}$",
        f"{detector_number}",
        f"{res['avg_hw_mean']:.2f}",
        f"{res['max_hw_mean']:.2f}",
        f"{res['ratio']:.4f}"
    ])

latex_table = tabulate(
    table_data,
    headers=[
        "$[n,k,d]$",
        "$r$",
        "Syndrome Length",
        "Average HW (Mean)",
        "Max HW (Mean)",
        "Average HW / Syndrome Length"
    ],
    tablefmt="latex"
)

print(r"\begin{table}[h]")
print(r"\centering")
print(r"\caption{Average and maximum syndrome Hamming weights and their ratio to syndrome length for different $[n,k,d]$ codes with various $r$.}")
print(r"\label{tab:avg_max_hw_ratio}")
print(latex_table)
print(r"\end{table}")
```
