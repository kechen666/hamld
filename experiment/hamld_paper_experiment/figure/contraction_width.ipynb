### EPMLD 相对于MLD的优化
（实际执行时间，时间成本，空间成本）
1. 采用order的情况下，随着码矩增加，优化的效果。
2. 采用order的情况下，随着重复轮次增加，优化的效果。
3. 采用order+slice的情况下，随着slice边数量的增加，性能的变换。

由于时间的执行时间，MLD只能够执行完d=5,r=3，其他的规模，MLD方法完全无法短时间运行，而且非常可能超出内存限制。

对于MLD的具体解码运行，我们只考虑运行d<=5, r<=3的情况。

其他情况下MLD，我们只进行估计其运行的时间成本，空间成本。

1. 首先先估计上述三种情况下的时间成本和空间成本。
2. 第二，我们需要估计EPMLD的实际运行时间和时间成本、空间成本的情况。

```python
from epmld import ContractionStrategy
from epmld.benchmark.contraction_strategy_benchmarking import get_circuit_by_stim

acdp: float = 0.001
arfp: float = 0.0001
bmfp: float = 0.005
brddp: float = 0.001

# 初始化存储不同策略时间开销和宽度的列表
mld_contraction_time_cost = []
mld_contraction_width = []
order_contraction_time_cost = []
order_contraction_width = []  # 修正变量名重复
slice_contraction_time_cost = []
slice_contraction_width = []  # 新增存储宽度
slice_update_contraction_time_cost = []
slice_update_contraction_width = []  # 新增存储宽度

# 遍历不同的d和r值
for d in [3, 5, 7, 9]:
    for r in [1, 2, 3, 4, 5]:
        # 获取电路和相关的错误模型
        circuit = get_circuit_by_stim(d=d, r=r, acdp=acdp, arfp=arfp, bmfp=bmfp, brddp=brddp)
        detector_error_model = circuit.detector_error_model(flatten_loops=True)

        # MLD策略
        mld_contraction_strategy = ContractionStrategy()
        mld_contraction_strategy.get_contraction_strategy(
            detector_error_model=detector_error_model,
            order_strategy="mld",
            perform_slicing=False
        )
        mld_contraction_time_cost.append(mld_contraction_strategy.contraction_cost)
        mld_contraction_width.append(mld_contraction_strategy.contraction_width)

        # 贪婪顺序策略 (Order)
        order_contraction_strategy = ContractionStrategy()
        order_contraction_strategy.get_contraction_strategy(
            detector_error_model=detector_error_model,
            order_strategy="greedy",
            perform_slicing=False
        )
        order_contraction_time_cost.append(order_contraction_strategy.contraction_cost)
        order_contraction_width.append(order_contraction_strategy.contraction_width)

        # 贪婪顺序 + 切片策略 (Slice)
        slice_contraction_strategy = ContractionStrategy()
        slice_contraction_strategy.get_contraction_strategy(
            detector_error_model=detector_error_model,
            order_strategy="greedy",
            perform_slicing=True,
            slice_strategy="parallelism",
            sliced_hyperedge_count=8,
            update_order=False
        )
        slice_contraction_time_cost.append(slice_contraction_strategy.sliced_contraction_cost)
        slice_contraction_width.append(slice_contraction_strategy.sliced_contraction_width)

        # 贪婪顺序 + 切片 + 更新顺序策略 (Slice + Update Order)
        slice_update_contraction_strategy = ContractionStrategy()
        slice_update_contraction_strategy.get_contraction_strategy(
            detector_error_model=detector_error_model,
            order_strategy="greedy",
            perform_slicing=True,
            slice_strategy="parallelism",
            sliced_hyperedge_count=8,
            update_order=True
        )
        slice_update_contraction_time_cost.append(slice_update_contraction_strategy.sliced_contraction_cost)
        slice_update_contraction_width.append(slice_update_contraction_strategy.sliced_contraction_width)

# 输出结果
print("MLD Contraction Time Costs:", mld_contraction_time_cost)
print("MLD Contraction Widths:", mld_contraction_width)
print("Order Contraction Time Costs:", order_contraction_time_cost)
print("Order Contraction Widths:", order_contraction_width)
print("Slice Contraction Time Costs:", slice_contraction_time_cost)
print("Slice Contraction Widths:", slice_contraction_width)
print("Slice Update Contraction Time Costs:", slice_update_contraction_time_cost)
print("Slice Update Contraction Widths:", slice_update_contraction_width)
```
